# CRS Data Loader v2

Simple 4-phase workflow for loading CRS master data:
1. **Tenant & Branding** - Organization setup and UI customization
2. **Boundaries** - Administrative hierarchy (State → District → Block)
3. **Common Masters** - Departments, designations, complaint types
4. **Employees** - Staff accounts with roles and assignments

---

```python
# Cell 1: Setup
from crs_loader import CRSLoader
```

```python
# Cell 2: Connect & Login
# Change URL to your DIGIT environment
loader = CRSLoader("https://unified-dev.digit.org")
loader.login()
```

```python
# Cell 3: Phase 1 - Tenant & Branding
loader.load_tenant("templates/Tenant And Branding Master.xlsx")
```

```python
# Cell 4: Phase 2 - Boundaries
# Note: Use your boundary Excel file (not in templates folder)
loader.load_boundaries("Boundary Master.xlsx", hierarchy_type="ADMIN")
```

```python
# Cell 5: Phase 3 - Common Masters
loader.load_common_masters("templates/Common and Complaint Master.xlsx")
```

```python
# Cell 6: Phase 4 - Employees
loader.load_employees("templates/Employee_Master_Dynamic_statea.xlsx")
```
